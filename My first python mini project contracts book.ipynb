```python
contacts = {}

def add_contact():
    name = input("Enter a name")
    phone = input("Enter a phone number")
    contacts[name] = phone
    print("contact added!")

def view_contacts():
    if not contacts:
        print("No contracts saved")
    else:
        print("All Contacts")
        for name,phone in contacts.items():
            print(name,":",phone)

def search_contact():
    name = input("Enter name to serach:")
    if name in contacts:
        print("phone number:",contacts[name])
    else:
        print("Contract not found")     


def delete_contact():
    name = input("Enter name to delete")
    if name in contacts:
        del contacts[name]
        print("Contact deleted")
    else:
        print("Not found")

            
while True:
    print("1.Add 2.view 3.Search 4.Delete")
    choice = input("Enter a choice")

    if choice == "1":
        add_contact()
    elif choice == "2":
        view_contacts()
    elif choice == "3":
        search_contact()
    elif choice == "4":
        delete_contact()
    elif choice == "5":
        print("Goodbye Bharath ")
        break
    else:
        print("Invalid choice")
            





        
      
```
