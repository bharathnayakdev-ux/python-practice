```python
nums = [1, 2, 3, 4]

result = list(map(lambda x: x * 3, nums))

print(result)
```

```python
square = lambda x: x * x

print(square(7))
```

```python
add3 = lambda a,b,c: a + b + c
print(add3(2,3,4))
```

```python
nums = [1,4,7,8,10,13]
result = list(filter(lambda x: x % 2 == 0 ,nums))
print(result)
```
