```python
def palindrome(s):

    reverse = ""

    for ch in s:
        reverse = ch + reverse

    if s == reverse:
        return True
    else:
        return False


print(palindrome("hello"))    

        
```

```python
def palindrome(s):
    s = s.lower()
    reverse = ""
    s = s.replace(" ","")

    for ch in s:
        reverse = ch + reverse

    return s == reverse
print(palindrome("Racec ar"))
```

```python
def string(t):
    t = t.lower()
    vowels_letters = "aeiou"
    constants = 0
    vowels = 0
    

    for ch in t:
        if ch.isalpha():
            if ch in vowels_letters:
                vowels += 1
            else:
                constants += 1

    print("vowels:",vowels)
    print("consonats:",constants)

string("Hello")                

        
```
