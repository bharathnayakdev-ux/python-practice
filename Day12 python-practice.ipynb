```python
def get_key(d):
    keys_list = list(d.keys())
    result = []

    i = 0
    while i < len(keys_list):
        result.append(keys_list[i])
        i += 1
    return result

print(get_key({"x": 5, "y": 9, "z": 2}))    


```

```python
def add(t):

    total = 0

    for i in t:
        total += i
    return total

print(add((1, 2, 3, 4)))    
```

```python
def odd(lst):
    odds = []

    for i in lst:
        if i % 2 != 0:
            odds.append(i)
    return odds

print(odd([2,7,89,34,65,75,4]))
        
            
        
```
