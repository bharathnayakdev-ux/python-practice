```python
def first_element(lst):
    return lst[0]

print(first_element([10,34,56]))
```

```python
def last_element(lst):
    return lst[-1]

print(last_element([5,9,2]))
```

```python
def list_sum(lst):
    total = 0
    for i in lst:
        total += i
    return total

print(list_sum([1, 2, 3, 4]))    
```

```python
def list_even(lst):
    count = 0
    for i in lst:
        if i % 2 == 0:
            count += 1

    return count
print(list_even([1, 2, 3, 4, 6]))        
```

```python
def list_max(lst):
    max_num = lst[0]

    for i in lst:
        if i > max_num:
           max_num = i
            
    return max_num
print(list_max([3, 8, 1, 10]))        


            
            
            
```

```python
def list_min(lst):
    min_num = lst[0]

    for i in lst:
        if i < min_num:
            min_num = i
    return min_num
print(list_min([3, 8, 1, 10]))        
        
```

```python
def second_largest(lst):
    temp = lst.copy()
    largest = max(temp)
    temp.remove(largest)
    return max(temp)

print(second_largest([3, 8, 1, 10]))
```

```python
def even_numbers(lst):
    evens = []

    for i in lst:
        if i % 2 == 0:
            evens.append(i)
    return evens

print(even_numbers([1, 2, 3, 4, 6]))
  
```

```python
def double_list(lst):
    new_list = []

    for i in lst:
        new_list.append(i * 2)
    return new_list

print(double_list([1, 2, 3]))
```
