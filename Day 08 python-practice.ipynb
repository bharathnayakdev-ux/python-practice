```python
def element(lst):
    elements = []
    i = 0
    while i < len(lst):
        elements.append(lst[i])
        i += 1
    return elements

print(element([10, 20, 30, 40]))    

```

```python
def add(lst):
    total = 0
    i = 0

    while i < len(lst):
        total += lst[i]
        i += 1
    return total

print(sum([1, 2, 3, 4]))    



```

```python
def even(lst):
    even_count = 0
    i = 0
    while i < len(lst):
        if lst[i] % 2 == 0:
            even_count += 1
        i += 1
    return even_count

print(even([1, 2, 4, 7, 8]))        
```

```python
def largest(lst):
    largest_num = lst[0]
    i = 0
    while i < len(lst):
        if lst[i] > largest_num:
            largest_num = lst[i]
        i += 1
    return largest_num

print(largest([3, 8, 1, 10]))         
```

```python
def postive_number(lst):
    postive_numbers = []
    i = 0

    while i < len(lst):
        if lst[i] > 0:
            postive_numbers.append(lst[i])

        
        i += 1
        
    return postive_numbers

print(postive_number([-2, 5, -1, 0, 7]))        

```

```python
def reverse_number(lst):
    reverse_numbers = []
    i = len(lst) -1
    while i >= 0:
        reverse_numbers.append(lst[i]) 
        i -= 1
    return reverse_numbers

print(reverse_number([10, 20, 30, 40]))    
```

```python
def reverse(lst):
    reverse_numbers = []

    for i in range(len(lst) -1,-1,-1):
        reverse_numbers.append(lst[i])
    return reverse_numbers

 
print(reverse([5, 15, 25]))

```

```python
def only_vowels(text):
    vowels = "aeiou"
    result = ""

    for ch in text:
        if ch in vowels:
            result += ch
    return result

print(only_vowels("bharath"))
print(only_vowels("hello"))        


```

```python
def only_consonants(text):
    vowels = "aeiou"
    result = ""

    for ch in text:
        if ch not in vowels:
            result += ch
    return result

print(only_consonants("bharath")) 
print(only_consonants("hello"))               
```

```python
def digits(text):
    result = ""

    for ch in text:
        if ch.isdigit():
            result += ch
    return result

print(digits("Bharath123"))        
```

```python
def string(text):
    result = ""

    for ch in text[::-1]:
            result += ch
    return result

print(string("bharath"))        

```

```python
def characters(text):
    result = ""

    for ch in text:
        if ch.isupper():
            result += ch
    return result

print(characters("BhaRathGPT"))        
```
