```python
def even(lst):
    evens = []
    for i in lst:
        if i % 2 == 0:
            evens.append(i)
    return evens

print(even([1,2,3,4,6]))  
```

```python
def odd(lst):
    odds = []
    for i in lst:
        if i % 2 != 0:
            odds.append(i)
    return odds
print(odd([1,2,3,4,6]))        
```

```python
def greater_number(lst):
    greater_numbers = []
    for i in lst:
        if i > 5:
            greater_numbers.append(i)
    return greater_numbers
print(greater_number([1, 6, 3, 10, 4]))        
```

```python
def postive_number(lst):
    postive_numbers = []
    for i in lst:
        if i > 0:
            postive_numbers.append(i)
    return postive_numbers

print(postive_number([-2, 5, -1, 0, 7]))

```

```python
def multiple(lst):
    multiple_of_three = []

    for i in lst:
        if i % 3 == 0:
            multiple_of_three.append(i)
    return multiple_of_three

print(multiple([1, 3, 4, 6, 9, 10]))        
```

```python
def double(lst):
    doubled = []

    for i in lst:
        doubled.append(i*2)
    return doubled

print(double([1,2,3]))        
```

```python
def square(lst):
    squares = []

    for i in lst:
        squares.append(i * i)
    return squares 

print(square([1, 2, 3, 4]))   
```

```python
def increased(lst):

    increased_by_10 = []

    for i in lst:
        increased_by_10.append(i + 10)
    return increased_by_10

print(increased([1, 2, 3]))    
```

```python
def negative(lst):

    negative_numbers = []

    for i in lst:
        negative_numbers.append(-i)
    return negative_numbers

print(negative([5, -2, 7]))    
```
