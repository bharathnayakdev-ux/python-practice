```python
prices = {
    "apple":100,
    "banana":40,
    "mango":80,
    "orange":60
}

count = 0 

for key,value in prices.items():
    if value > 70:
        print(key,"=",value)
        count += 1
      
print("count =",count)        
```

```python
nums = [2,5,8,10]

count = 0
i = 0

while i < len(nums):
    if nums[i] > 6:
        count += 1
    i += 1
print("count =",count)       
```

```python
list = [1,2,3,4,5]

total = 0
i = 0

while i < len(list):
    total += list[i]
    i += 1

print("sum =",total)    
    
```

```python
nums = [10,15,20,25]

count = 0
i = 0

while i < len(nums):
    if nums[i] % 5 == 0:
        count += 1
    i += 1
print("count =",count)        
```
