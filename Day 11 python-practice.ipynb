```python
def first_element(t):
    return t[0]

print(first_element((5,9,2)))
```

```python
def last_element(t):
    return t[-1]

print(last_element((5, 9, 2)))
```

```python
def add(t):

    total = 0

    for i in t:
        total += i
    return total

print(add((1, 2, 3, 4)))    
```

```python
def unquie_elements(lst):
    return set(lst)

print(unquie_elements([1, 2, 2, 3, 3, 4]))
```

```python
def union(set1,set2):
    return set1 | set2

print(union({1, 2, 3}, {3, 4, 5}))
```

```python
def common_elements(set1,set2):
    return set1 & set2

print(common_elements({1,2,3},{2,3,4}))



```
