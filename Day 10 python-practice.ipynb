```python
def add(d):
    total = 0

    for value in d.values():
        total += value

    return total

print(add({"math": 80, "english": 70, "science": 90}))    

```

```python
def highest_marks(d):
    highest_subject = ""
    highest_score = 0

    for subject,score in d.items():
        if score > highest_score:
            highest_score = score
            highest_subject = subject

    return highest_subject

print(highest_marks({"math": 80, "english": 70, "science": 90}))        
        





```

```python
def subject(d):
    
    subjects = []

    for subject,marks in d.items():
        if marks > 75:
            subjects.append(subject)
    return subjects

print(subject({"math": 80, "english": 70, "science": 90}))        


```

```python
def add(d):
    values = list(d.values())
    total = 0

    i = 0
    while i < len(values):
        total += values[i]
        i += 1

    return total

print(add({"a": 10, "b": 20, "c": 30}))
```

```python
def get_key(d):
    keys_list = list(d.keys())
    result = []

    i = 0
    while i < len(keys_list):
        result.append(keys_list[i])
        i += 1
    return result

print(get_key({"x": 5, "y": 9, "z": 2}))    


```
