```python
def vowel_count(text):
    vowels = "aeiou"
    count = 0
    i = 0

    while i < len(text):
        if text[i] in vowels:
            count += 1
        i += 1
    return count        
    
        
print(vowel_count("bharath"))        
```

```python
def reverse_string(text):
    result = ""
    i = len(text) - 1

    while i >= 0:
        result += text[i]
        i -= 1

    return result

print(reverse_string("hello"))
```

```python
def digits(text):
    count = 0
    i = 0

    while i < len(text):
        if text[i].isdigit():
            count += 1
        i += 1
    return count

print(digits("Bharath123"))        


```

```python
def string(text):
    count = 0
    i = 0

    while i < len(text):
        if text[i] in "a":
            count += 1
        i += 1
    return count

print(string("bharath"))        
```

```python
def string(text):
    result = ""
    i = 0

    while i < len(text):
        if text[i].isupper():
            result += text[i]
        i += 1
    return result

print(string("BhaRatjGPT"))        
```
